# 4 Analyzing Null Values 

To check for null values in a pandas DataFrame, one can use the isnull()
method,which returns a DataFrame of the same shape as the original, 
where each cell contains a Boolean value indicating whether the corresponding 
cell in the original DataFrame is null or not. To count the number of null values, 
sum() function is used, which counts True values (True represents a null value).


```python
df.isnull()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

The purpose of this code is to provide a  summary of missing values in each column of the DataFrame (df).
It calculates the count of missing values for each column, sorts these counts in descending order, and presents
the information in a clear manner, allowing for easy identification of columns with the most missing data. 
This information is vital for prioritizing data cleaning and imputation efforts during the data preprocessing
stage and aids in understanding the quality  of the dataset in the context of exploratory data analysis (EDA).

```python
print(df.nunique().sort_values())
print(df.info())
```

In summary, the code first sorts and prints the DataFrame's column names based on the number of unique values in each column and then prints the summary information about the DataFrame using df.info(). This is useful for understanding the data and identifying any potential data quality issues, especially if you're working with a large or unfamiliar dataset.

```python
df.groupby ("Outcome").mean()
```

The code df.groupby("Outcome").mean() is used to group a DataFrame df by a column named "Outcome"
and calculate the mean (average) values for each group within that column. This is typically used
in data analysis to find the average of other columns' values for each unique outcome category in
the "Outcome" column.

```python
df.groupby ("Outcome").agg(["mean","median"])
```

The code df.groupby("Outcome").agg(["mean", "median"]) is used to group a DataFrame df by a column named "Outcome"
and then calculate both the mean and median for each group within that column. This code provides summary statistics
for each unique outcome category in the "Outcome" column, showing both the average (mean) and the middle 
value (median) of the other columns for each group.

# Analysis of Glucose And Outcome

```python
fig15, ax15 = plt.subplots(figsize=(20,8))
sns.histplot(data = df ,x="Glucose", hue= "Outcome",shrink= 0.8, multiple='stack',kde=True,ax=ax15)
plt.show()
```

The provided code is using the matplotlib library for creating a single-figure plot and the seaborn library to create a stacked histogram with KDE (Kernel Density Estimate) for the "Glucose" column in a DataFrame df

-First line creates a matplotlib figure with a single subplot.
-Second line uses seaborn (sns) to create a histogram and KDE plot for the "Glucose" column of the DataFrame df.
-The hue parameter is used to differentiate the data points based on the "Outcome" column.
-The shrink parameter controls the width of the bars in the histogram. A value less than 1 (0.8 in this case)
 makesThe bars narrower, and they are stacked closely together.
-The multiple parameter is set to 'stack', which means that the histograms for different "Outcome" categories are       stacked on top of each other.
This is a useful visualization for understanding the distribution of glucose levels in the context of different outcomes, such as diabetes or non-diabetes.

# 5 Filling missing values

```python
# first copy the original dataframe into a new pandas dataframe
df_cp = df.copy()
# now fill the missing values with the median value of the perticular columns
df_cp["Glucose"] = df ["Glucose"].replace(0,df["Glucose"].median())
df_cp["BloodPressure"] = df ['BloodPressure'].replace(0,df["BloodPressure"].median())
df_cp["SkinThickness"] = df ['SkinThickness'].replace(0,df["SkinThickness"].median())
df_cp["Insulin"] = df ["Insulin"].replace(0,df["Insulin"].median())
df_cp["BMI"] = df ['BMI'].replace(0,df["BMI"].median())


```

```python
df_cp.describe().T
```

```python
corr=df_cp.corr()
corr
```

```python
plt.figure(figsize=(10,5))

sns.heatmap(corr[(corr>=0.5) | (corr <= -0.5)],
           cmap='YlGnBu',vmax=1.0,vmin=-1.0,
           annot=True,annot_kws={'size':15})

plt.title('Correlation between Variables')
plt.show()
```
